```python
# Function to find the percentage of Null Values present in a data set.

def Train_missing_values(df):
    Missing_Data_Percent=pd.DataFrame(df.isna().sum())
    Missing_Data_Percent.reset_index(inplace=True)
    Missing_Data_Percent.columns=['Feild_Name','Missing_value_count']
    Missing_Data_Percent['Percent_missing_values']=Missing_Data_Percent['Missing_value_count'].\
                                                apply(lambda Missing_value_count:(Missing_value_count/len(df))*100)
    return Missing_Data_Percent.sort_values(['Percent_missing_values'],ascending=False)


Train_missing_values(df)
```

# K Nearest Neighbor(KNN)


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Training set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
# Modeling the KNN Classifier

from sklearn.neighbors import KNeighborsClassifier
k = 3
#Train Model and Predict  
kNN_model = KNeighborsClassifier(n_neighbors=k).fit(X_train,y_train)
kNN_model

yhat = kNN_model.predict(X_test)
yhat[0:5]
```

```python
# Finding the Best K

Ks=15
mean_accuracy=np.zeros((Ks-1))
std_accuracy=np.zeros((Ks-1))
ConfustionMx=[];
for n in range(1,Ks):
    
    #Train the Model and Predict  
    
    KNN = KNeighborsClassifier(n_neighbors=n).fit(X_train,y_train)
    yhat = KNN.predict(X_test)
    
    
    mean_accuracy[n-1]=np.mean(yhat==y_test);
    
    std_accuracy[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])
mean_accuracy
```

```python
# Building the model again, using k=7

from sklearn.neighbors import KNeighborsClassifier
k = 7
#Train Model and Predict  

KNN = KNeighborsClassifier(n_neighbors=k).fit(X_train,y_train)
KNN
```

# Classification Models Evaluation 

```python
from sklearn.metrics import jaccard_similarity_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
```

# Jaccard Index

```python
# Evaluate KNN
knn_yh = KNN.predict(X_test)

jc1 = round(jaccard_similarity_score(y_test, knn_yh),2)

# Evaluate the Decision Trees

dt_yh = DT.predict(X_test)
jc2 = round(jaccard_similarity_score(y_test, dt_yh),2)

# Evaluate SVM

svm_yh = SVM.predict(X_test)
jc3 = round(jaccard_similarity_score(y_test, svm_yh),2)

# Evaluate Logistic Regression

lr_yh = Logr.predict(X_test)
jc4 = round(jaccard_similarity_score(y_test, lr_yh),2)

list_jc = [jc1, jc2, jc3, jc4]
list_jc
```

# F1 Score

```python
# Evaluate KNN

fs1 = round(f1_score(y_test, knn_yh, average='weighted'), 2)

# Evaluate the Desision Trees 

fs2 = round(f1_score(y_test, dt_yh, average='weighted'), 2)

# Evaluate the SVM Model

fs3 = round(f1_score(y_test, svm_yh, average='weighted'), 2)

# evaluate Logistic Regression
fs4 = round(f1_score(y_test, lr_yh, average='weighted'),2 )

list_fs = [fs1, fs2, fs3, fs4]
list_fs
```

# Log loss

```python
# Log Loss closely related to cross entrophy

from sklearn.metrics import log_loss
lr_prob = Logr.predict_proba(X_test)
LR_yh_prob = Logr.predict_proba(X_test)

list_ll = ['NA', 'NA', 'NA', round(log_loss(y_test, LR_yh_prob), 2)]
list_ll
```

```python
import pandas as pd

# fomulate the report format
# Create a crosstab table

df = pd.DataFrame(list_jc, index=['KNN','Decision Tree','SVM','Logistic Regression'])
df.columns = ['Jaccard']
df.insert(loc=1, column='F1-score', value=list_fs)
df.insert(loc=2, column='LogLoss', value=list_ll)
df.columns.name = 'Algorithm'
df
```
